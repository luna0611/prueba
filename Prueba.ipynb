#### **Outliers**

En cuanto a los valores atípicos se puede observar en las variables que se presentan los siguientes hallazgos:

```python
#import seaborn as sns
#import matplotlib.pyplot as plt

#sns.boxplot(data=datac['vble0'])
```

```python
#df=datac[['vble0', 'vble2' , 'vble5', 'vble8']]
df=datac
```

```python
def get_iqr_values(df_in, col_name):
    median = df_in[col_name].median()
    q1 = df_in[col_name].quantile(0.25) # 25th percentile / 1st quartile
    q3 = df_in[col_name].quantile(0.75) # 7th percentile / 3rd quartile
    iqr = q3-q1 #Interquartile range
    minimum  = q1-1.5*iqr # The minimum value or the |- marker in the box plot
    maximum = q3+1.5*iqr # The maximum value or the -| marker in the box plot
    return median, q1, q3, iqr, minimum, maximum

def get_iqr_text(df_in, col_name):
    median, q1, q3, iqr, minimum, maximum = get_iqr_values(df_in, col_name)
    text = f"median={median:.2f}, q1={q1:.2f}, q3={q3:.2f}, iqr={iqr:.2f}, minimum={minimum:.2f}, maximum={maximum:.2f}"
    return text

def remove_outliers(df_in, col_name):
    _, _, _, _, minimum, maximum = get_iqr_values(df_in, col_name)
    df_out = df_in.loc[(df_in[col_name] > minimum) & (df_in[col_name] < maximum)]
    return df_out

def count_outliers(df_in, col_name):
    _, _, _, _, minimum, maximum = get_iqr_values(df_in, col_name)
    df_outliers = df_in.loc[(df_in[col_name] <= minimum) | (df_in[col_name] >= maximum)]
    return df_outliers.shape[0]

def box_and_whisker(df_in, col_name):
    title = get_iqr_text(df_in, col_name)
    sns.boxplot(df_in[col_name])
    plt.title(title)
    plt.show()
```

Visualizando los datos

```python
box_and_whisker(df, 'vble0')
box_and_whisker(df, 'vble2')
box_and_whisker(df, 'vble5')
box_and_whisker(df, 'vble8')

_, _, _, _, minimum_vble2_before, maximum_vble2_before = get_iqr_values(df, 'vble5')

# %% Count the outliers in the original data frame
print(f"vble0 has {count_outliers(df, 'vble0')} outliers")
print(f"vble2 has {count_outliers(df, 'vble2')} outliers")
print(f"vble5 has {count_outliers(df, 'vble5')} outliers")
print(f"vble8 has {count_outliers(df, 'vble8')} outliers")
```

Eliminar los valores atípicos y visualizar el resultado

```python
# %% eliminar outliers
print(f"rows before removing: {df.shape[0]}")
df = remove_outliers(df, 'vble0')
df = remove_outliers(df, 'vble2')
df = remove_outliers(df, 'vble5')
df = remove_outliers(df, 'vble8')
print(f"rows after removing: {df.shape[0]}")

# visualización si outliers
box_and_whisker(df, 'vble0')
box_and_whisker(df, 'vble2')
box_and_whisker(df, 'vble5')
box_and_whisker(df, 'vble8')

_, _, _, _, minimum_vble0_after, maximum_vble0_after = get_iqr_values(df, 'vble5')
```

#### **Imputación de variables**


**Imputación de variables númericas**


se utiliza para la imputación de variables númericas la “mediana” como estrategia de limpieza.

```python
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
# Set up the imputer
median_imputer = SimpleImputer(missing_values=np.nan, strategy='median')
 
# fit the imputer
median_imputer.fit(datac[['vble2', 'vble5', 'vble8']])

# transform the data
data_t = median_imputer.transform(datac[['vble2', 'vble5', 'vble8']])
```

```python
# Conteo relativo de valores por columna de no nulos
round((data_t.count()/len(data_t))*100,0).astype("int")
```

**Imputación de variables categóricas**
